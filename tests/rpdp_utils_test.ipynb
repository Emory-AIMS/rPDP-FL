# Demo for simulating different distributions

```python
import sys
sys.path.append("..")
from fedrpdp.utils.rpdp_utils import MultiLevels, MixGauss, Pareto

GENERATE_EPSILONS = {
    "TwoLevels": lambda n, params: MultiLevels(2, *params, n),
    "ThreeLevels": lambda n, params: MultiLevels(3, *params, n),
    "BoundedPareto": lambda n, params: Pareto(*params, n), 
    "BoundedMixGauss": lambda n, params: MixGauss(*params, n),
}
SETTINGS = {
    "TwoLevels": [[[0.8,0.2], [0.5, 5.0]]],
    "ThreeLevels": [[[0.7,0.2,0.1], [0.5, 2.0, 5.0]]],
    "BoundedPareto": [[1.0, 0.5]], 
    "BoundedMixGauss": [[[0.7,0.2,0.1], [(0.5, 0.1), (2.0, 0.5), (5.0, 1.0)]]],
}
bounded_func = lambda values, lower, upper: [min(max(x, lower), upper) for x in values]
```

## Test random distribution of `TwoLevels` and `ThreeLevels`

```python
import numpy as np

target_epsilons = [GENERATE_EPSILONS["TwoLevels"](6000, params) for params in SETTINGS["TwoLevels"]]
print(target_epsilons[0])
print('Num of records for epsilon=0.5: ', np.sum(target_epsilons[0] == 0.5))
print('Num of records for epsilon=5.0: ', np.sum(target_epsilons[0] == 5.0))

target_epsilons = [GENERATE_EPSILONS["ThreeLevels"](6000, params) for params in SETTINGS["ThreeLevels"]]
print(target_epsilons[0])
print('Num of records for epsilon=0.5: ', np.sum(target_epsilons[0] == 0.5))
print('Num of records for epsilon=2.0: ', np.sum(target_epsilons[0] == 2.0))
print('Num of records for epsilon=5.0: ', np.sum(target_epsilons[0] == 5.0))
```

## Test random distribution of `BoundedPareto` and `BoundedMixGauss`

```python
lower_eps, upper_eps = 0.1, 10.0
target_epsilons = [GENERATE_EPSILONS["BoundedMixGauss"](6000, params) for params in SETTINGS["BoundedMixGauss"]]
target_epsilons = [bounded_func(element, lower_eps, upper_eps) for element in target_epsilons]
```

```python
# Vision
import matplotlib.pyplot as plt
import seaborn as sns
plt.close('all')
plt.rcParams['font.sans-serif'] = ['SimSun']
legend_font = {
    'style': 'normal',
    'size': 8,  # 字号
    'weight': "normal",  # 是否加粗，不加粗
}
sns.set_style('whitegrid', {'axes.linewidth': 1, 'axes.edgecolor':'black'}) #轮廓线
sns.set_palette(palette=sns.color_palette('bright')) #颜色

fig, ax = plt.subplots(1, 1, figsize=(4, 3), constrained_layout=True)

for i, params in enumerate(SETTINGS["BoundedMixGauss"]):
    sns.histplot(x=target_epsilons[i], ax=ax, label=f'means={params[1][0][0]}/{params[1][1][0]}/{params[1][2][0]}\nstds={params[1][0][1]}/{params[1][1][1]}/{params[1][2][1]}')

ax.set_ylabel('Density', fontsize=8)
ax.set_xlabel('Epsilon', fontsize=8)
ax.tick_params(labelsize=8)
ax.set_ylim(0, 2000)
ax.set_xlim(0, 10.2)
ax.legend(loc='upper right', prop=legend_font)
ax.grid(True)

# ax.set_title(f'N={60000}', fontsize=9)
fig.suptitle(f'N={60000}', fontsize=9)
plt.show()
```

```python
SETTINGS["BoundedPareto"]=[[1.0, 0.5], [1.5, 0.5], [2.5, 0.5]]
target_epsilons = [GENERATE_EPSILONS["BoundedPareto"](6000, params) for params in SETTINGS["BoundedPareto"]]
target_epsilons = [bounded_func(element, lower_eps, upper_eps) for element in target_epsilons]
```

```python
# Vision
import matplotlib.pyplot as plt
import seaborn as sns
plt.close('all')
plt.rcParams['font.sans-serif'] = ['SimSun']
legend_font = {
    'style': 'normal',
    'size': 8,  # 字号
    'weight': "normal",  # 是否加粗，不加粗
}
sns.set_style('whitegrid', {'axes.linewidth': 1, 'axes.edgecolor':'black'}) #轮廓线
sns.set_palette(palette=sns.color_palette('bright')) #颜色

fig, ax = plt.subplots(1, 1, figsize=(4, 3), constrained_layout=True)

colors = ['b','g','grey']
for i, params in enumerate(SETTINGS["BoundedPareto"]):
    sns.histplot(x=target_epsilons[i], ax=ax, color=colors[i], label=f'shape={params[0]}, lower={params[1]}')

ax.set_ylabel('Density', fontsize=8)
ax.set_xlabel('Epsilon', fontsize=8)
ax.tick_params(labelsize=8)
ax.set_xlim(0, 10.2)
ax.legend(loc='upper right', prop=legend_font)
ax.grid(True)

# ax.set_title(f'N={6000}', fontsize=9)
fig.suptitle(f'N=6000', fontsize=9)
plt.show()
```
